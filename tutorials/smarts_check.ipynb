```python
from chython import smarts
from chython.exceptions import IncorrectSmiles

from tqdm.auto import tqdm
```

```python
# pathw to the rule files that extracted under same configuration
tsv_path_a = "../mapping/uspto_full_rules_light_pop_3_uu.tsv"
tsv_path_b = "../mapping/uspto_full_rules_light_pop_3_uuu.tsv"
```

```python
def load_rules(tsv_path):
    data_q = []
    errors = []
    with open(tsv_path, encoding="utf-8") as f:
        f.readline()  # skip header
        for line in tqdm(f):
            line = line.rstrip("\n")
            if not line:
                continue
            try:
                rule_smarts = line.split("\t", 1)[0]
                q = smarts(rule_smarts)
                cq = ~q
                data_q.append((q, cq))
            except IncorrectSmiles:
                errors.append(rule_smarts)
    return data_q, errors
```

```python
a_q, errors_a = load_rules(tsv_path_a)
```

```python
b_q, errors_b = load_rules(tsv_path_b)
```

```python
difference_ab = list(set([cq for _, cq in a_q]) - set([cq for _, cq in b_q]))
print(len(difference_ab))
for r in difference_ab:
    r.clean2d()
    display(r)
    print(str(r))
```

```python
difference_ba = list(set([cq for _, cq in b_q]) - set([cq for _, cq in a_q]))
print(len(difference_ba))
for r in difference_ba:
    r.clean2d()
    display(r)
    print(str(r))
```

```python
# A \ B as ReactionContainer
b_cq_set = {cq for _, cq in b_q}
difference_ab_rxn = [q for q, cq in a_q if cq not in b_cq_set]

# B \ A as ReactionContainer (optional)
a_cq_set = {cq for _, cq in a_q}
difference_ba_rxn = [q for q, cq in b_q if cq not in a_cq_set]
```
